```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
sampling_rate = 1000  # Samples per second
duration = 1  # Duration in seconds
num_samples = int(sampling_rate * duration)
time = np.linspace(0, duration, num_samples, endpoint=False)

# Define the frequencies of the sine waves
f = ["10", "5", "3"]  # Hz
frequencies = [10, 5, 3]  # Hz

# Generate the sine waves
waves = [np.sin(2 * np.pi * freq * time) for freq in frequencies]



# Plot the sine waves in the time domain
for i, wave in enumerate(waves):
    resultant_wave += wave
    plt.plot(time, wave, label= f[i])
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

## Converting into frequency domain using FFT:

```python
# Perform FFT on each sine wave
fft_waves = [np.fft.fft(wave) for wave in waves]

# Get the amplitudes of the frequency components
amplitudes = [np.abs(fft_wave[:num_samples // 2]) for fft_wave in fft_waves]

# Calculate the corresponding frequencies
freqs = np.fft.fftfreq(num_samples, 1 / sampling_rate)[:num_samples // 2]

#print(freqs)

# Plot the amplitudes in the frequency domain
for i, amplitude in enumerate(amplitudes):
    plt.plot(freqs, amplitude, label=f[i])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

# Perform IFFT on each amplitude

```python

ifft_waves = [np.fft.ifft(fft_wave) for fft_wave in fft_waves]

# Plot the sine waves in the time domain
for i, wave in enumerate(ifft_waves):
    plt.plot(time, wave.real, label=f[i])
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```
