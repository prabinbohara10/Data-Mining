```python
df = pd.read_csv("../dataset/Dry_Bean_Dataset.csv")
```

```python
df_new =pd.get_dummies(df)
```

```python
df_new
```

```python
from sklearn import tree
```

```python
df_classifier = tree.DecisionTreeClassifier(criterion = "entropy", max_depth = 2)
```

```python
X = df_new.drop(columns= ['Play_No','Play_Yes'])
```

```python
X
```

```python
y = df_new.loc[:,'Play_No':'Play_Yes']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 12, test_size = 0.33)
```

```python
train = pd.concat([X_train, y_train] , axis = 1)
```

```python
X_train
```

```python
model = df_classifier.fit(X_train,y_train["Play_Yes"])
```

```python
X_test
```

```python
prediction = model.predict(X.loc[2:5,])
```

```python
prediction
```

```python
prediction_test = model.predict(X_test)
```

```python
prediction_test
```

```python
y_test["Play_Yes"]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test["Play_Yes"], prediction_test)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test["Play_Yes"], prediction_test)
print(report)
```
