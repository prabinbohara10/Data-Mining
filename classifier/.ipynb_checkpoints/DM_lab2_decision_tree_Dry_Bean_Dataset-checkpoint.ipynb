```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
main_df = pd.read_excel("../dataset/Dry_Bean_Dataset.xlsx")
```

```python
df = main_df
```

```python
# Import label encoder
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['Class_encode']= label_encoder.fit_transform(df['Class'])
df['Class_encode'].unique()
```

```python
df.head(10)
```

```python
sns.histplot(data = df["ConvexArea"], kde= True)
```

```python
X = df.iloc[:,:-2]
X
```

```python
y = df.loc[:,'Class_encode']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 100, test_size = 0.33)
```

```python
train = pd.concat([X_train, y_train] , axis = 1)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn import tree
df_classifier = tree.DecisionTreeClassifier()
```

```python
df_classifier.fit(X_train,y_train)
```

```python
prediction_test = df_classifier.predict(X_test)
```

```python
prediction_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, prediction_test)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, prediction_test)
print(report)
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(y_test, prediction_test)
cf_matrix
```

```python
labels = main_df["Class"].unique()
labels
```

```python
import seaborn as sns 
fig, ax = plt.subplots(figsize=(10,10))   
sns.heatmap(cf_matrix, annot=True, ax=ax, fmt ='g', linewidths = .5, xticklabels = labels, yticklabels = labels)
```

```python
#drawing the tree:
from sklearn.tree import plot_tree
plt.figure(figsize=(10, 8))
plot_tree(df_classifier)
```

```python
print(df_classifier.tree_.max_depth)
```

```python
feature_imp= df_classifier.feature_importances_ * 100
feautre_imp_cum = np.cumsum(df_classifier.feature_importances_ *100)
total_points =len(feature_imp)
```

```python
plt.plot(np.arange(total_points),feature_imp)
plt.plot(np.arange(total_points),feautre_imp_cum)
plt.title("Importance on various attibutes")
plt.xlabel("Feature Number")
plt.ylabel("Importance in percentage")
legend_class = ["Feature Importance", "Cummulative Feature Importance"]
plt.legend(labels = legend_class)
plt.show()
```

```python
path =df_classifier.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
```

```python
fig, ax = plt.subplots()
ax.plot(ccp_alphas[:-1], impurities[:-1], marker="o", drawstyle="steps-post")
ax.set_xlabel("effective alpha")
ax.set_ylabel("total impurity of leaves")
ax.set_title("Total Impurity vs effective alpha for training set")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
Scaled_X =  scaler.transform(X)
```

```python
from sklearn.decomposition import PCA
# Set the n_components=3
principal=PCA()
principal.fit(Scaled_X)
```

```python
explained_var_library = principal.explained_variance_ratio_ * 100
explained_var_cum_library = np.cumsum(principal.explained_variance_ratio_ * 100)
```

```python

total_points =len(explained_var_library)
```

```python
explained_var_library, explained_var_cum_library
```

```python
plt.plot(np.arange(total_points),explained_var_library)
plt.plot(np.arange(total_points),explained_var_cum_library)
plt.title("Scree Plot")
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance")
legend_class = ["Explained Variance", "Cummulative Explained Variance"]
plt.legend(labels = legend_class)
plt.show()
```

```python
principal=PCA(n_components = 7)
principal.fit(Scaled_X)
Y_library = principal.transform(Scaled_X)
```
