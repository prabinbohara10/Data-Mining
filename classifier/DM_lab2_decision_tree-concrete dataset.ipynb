```python
import pandas as pd
```

```python
main_df = pd.read_excel("../dataset/Concrete_Data.xlsx")
```

```python
df = main_df
```

```python
df.describe()
```

```python
df.columns
```

```python
df["Strength Class"] = pd.cut(df['Concrete compressive strength(MPa, megapascals) '],bins =3)
```

```python
df['Strength Class'].value_counts()
```

```python
# Import label encoder
from sklearn import preprocessing
  
label_encoder = preprocessing.LabelEncoder()
  

df['Strength Class']= label_encoder.fit_transform(df['Strength Class'])
  
df['Strength Class'].unique()
```

```python
df
```

```python
from sklearn import tree
```

```python
df_classifier = tree.DecisionTreeClassifier(criterion = "entropy")
```

```python
#SX = df.drop(columns= ['Strength Class'])
```

```python
X = df.iloc[:,:-2]
X
```

```python
y = df.loc[:,'Strength Class']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 100, test_size = 0.33)
```

```python
train = pd.concat([X_train, y_train] , axis = 1)
```

```python
X_train
```

```python
df_classifier.fit(X_train,y_train)
```

```python
X_test
```

```python
prediction =df_classifier.predict(X.loc[2:5,])
```

```python
prediction
```

```python
prediction_test = df_classifier.predict(X_test)
```

```python
prediction_test
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, prediction_test)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, prediction_test)
print(report)
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(y_test, prediction_test)
```

```python
import seaborn as sns 
fig, ax = plt.subplots(figsize=(10,10))   
sns.heatmap(cf_matrix, annot=True, ax=ax, fmt ='g', linewidths = .5)
```
